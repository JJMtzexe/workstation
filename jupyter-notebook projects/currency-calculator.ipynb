```python
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd

def getCurrencyTable():
    table = pd.read_html("https://developers.google.com/adsense/management/appendix/currencies")[0]
    table.columns = ["currency_code", "currency_name"]
    table.index += 1
    return table

def stringCleaner(string):
    sample = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ "
    new_string = ""
    for i in range(0, len(string)):
        if string[i] in sample:
            new_string += string[i]
    return new_string

def codeOutOfParenthesis(string):
    return string[int(string.find("(")+1):int(string.find("(")+4)]

def validateCountryCode(code, table):
    for x in table["currency_code"]:
        if x == code:
            return True
    return False

def getCountryNameFromCountryCode(code, table):
    for x in table["currency_code"]:
        if x == code:
            return table[table["currency_code"] == x]["currency_name"][table[table["currency_code"] == x]["currency_name"].index[0]]
    return False

def getCurrencyCoefficient():

    origin = input("Origin currency code: ").upper()
    if validateCountryCode(origin, getCurrencyTable()) == False:
        return Exception(f"\'{origin}\' is an invalid currency code, please try again \n You might also want to check the currency codes: https://developers.google.com/adsense/management/appendix/currencies")

    destiny = input("Destiny currency code: ").upper()
    if validateCountryCode(origin, getCurrencyTable()) == False:
        return Exception(f"\'{origin}\' is an invalid currency code, please try again. \n You might also want to check the currency codes: https://developers.google.com/adsense/management/appendix/currencies")

    currency_url = rq.get(f"https://www.google.com/finance/quote/{origin}-{destiny}?hl=es")
    convert_coefficient = BeautifulSoup(currency_url.text, 'html.parser')
    convert_coefficient = float(convert_coefficient.find_all("div", "YMlKec fxKbKc")[0].text.replace(",", "."))
    
    quantity = float(input("Quantity of " + getCountryNameFromCountryCode(origin, getCurrencyTable()) + "s" + " to convert to " + getCountryNameFromCountryCode(destiny, getCurrencyTable()) + "s:"))

    return f"{quantity} {getCountryNameFromCountryCode(origin, getCurrencyTable())}s = {round(quantity * convert_coefficient, 4)} {getCountryNameFromCountryCode(destiny, getCurrencyTable())}s"

print(getCurrencyCoefficient())
```
